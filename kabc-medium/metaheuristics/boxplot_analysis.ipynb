```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.reset_orig()
```

```python
from IPython.display import display

file_name = 'compa'
for function in ['Sphere', 'Rastrigin', 'Rosenbrock']:
    for inertia in ['decreasing', 'constant', 'clerc']:
        title = function + '_' + inertia
        df = pd.read_csv(title + '.csv', index_col=0)
        plt.figure(figsize=(8,6))
        data = pd.melt(df)
        data.columns = ['topologies', 'fitness']
        sns.boxplot(x='topologies', y='fitness', data=data, width=0.3)
        plt.xlabel('')
        plt.title(function.capitalize() + ' Function, inertia weight ' + inertia)
        plt.savefig(title + '_boxplot.jpg')
```

```python
from IPython.display import display
from pylab import rc

rc('axes', linewidth=0.5)
rc('xtick', labelsize=9)
rc('font', weight='light')

for function in ['Sphere', 'Rastrigin', 'Rosenbrock']:
    file_name = 'comparison_{}.csv'.replace('{}', function)
    df = pd.read_csv(file_name, index_col=0)
    df.columns = ['PSO', 'FSS', 'ABC']
    
    plt.figure(figsize=(8,6))
    data = pd.melt(df)
    
    sns.boxplot(x='variable', y='value', data=data, width=0.3, linewidth=0.7, color='w')
    plt.xlabel('')
    plt.ylabel('Fitness')
    plt.title('{} Function'.format(function))
    plt.savefig('figs/box_comparison_{}.png'.format(function.lower()))
```
